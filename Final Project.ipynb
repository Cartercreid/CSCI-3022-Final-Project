```python
#state pop data was pulled from the US 2015 Census
state_pop = [20271272,10214860,4896146,10042802,8382993,945934,6006401,1844128,672228,4425092,\
             6600299,4858979,2992333,2978204,4670724,3911338,27469114,8958013,12802503,19795791, \
            6590886,6794422,1330608,626042,1329328,11613423,6619680,9922576,12859995,5771337, \
            5489594,3123899,6083672,2911641,1896190,858469,756927,2085109,5456574,586107,\
             1032949,1654930,2995919,6828065,2890845,39144818,4028977,7170351,1431603,738432]

state_count = np.zeros(50)
state_rate = []
state_bins = []

for i in range(0,len(data)):
    state_count[data.state.iloc[i]-1] += 1
    
for i in range(0,len(state_count)):
    state_bins.append(i+1)
    state_rate.append(1000000 * (state_count[i] / state_pop[i]))

sdpm = pd.DataFrame({'Deaths by State':state_bins, 'Deaths/million residents':state_rate})
sdpm['Deaths by State'] = sdpm['Deaths by State'].map(r_state)

plt.figure(figsize=(15,8))
sns.barplot('Deaths by State', 'Deaths/million residents', data=sdpm, color="blue").set_title('South                               North East                      Mid West                                      West')
plt.axvline(16.5, 1, 0, color = 'black');
plt.axvline(24.5, 1, 0, color = 'black');
plt.axvline(36.5, 1, 0, color = 'black');

plt.axvline(8.5, .5, 0, color = 'red');
plt.axvline(12.5, .5, 0, color = 'r');
plt.axvline(19.5, .5, 0, color = 'r');
plt.axvline(29.5, .5, 0, color = 'r');
plt.axvline(44.5, .5, 0, color = 'r');
plt.show()
```

```python
"""One of our hypothesees was that southern states would have higher police caused fatalaties than their northern counterparts
In order to actually determine a rate, state population data was pulled form the US 2015 Census. Deaths/ Million Residents 
seemed like a good metric for a population normalized metric. The red bars separate the smaller geographic regions."""
```

```python
ax = sns.boxplot(x=state_rate).set_title("Rate/Million Residents - All States")
```

```python
"""Box plots are excellent becaause they encode many statistical metrics. The overall distribution for All states seems normal
except for one outlier (Oklahoma)"""
```

```python
#now lets look at the average rate by region

#South: 1 - 17
#South atlantic: 1-9
#East south Central: 10-13
#West South Central: 14-17

#North East: 18-25
#middle atlantic: 18 - 20
#New England: 20 - 25

#Mid-West: 26-37
#East North Central:26-30
#West North Central: 31-37

#West: 38-50
#Mountain: 38-45
#Pacific: 46-50

geo_bin = []
for i in range(1, len(state_rate)+1):
    if (i<=9):
        #South_Atlantic
        geo_bin.append(1)
    elif (i<=13):
        #East_South_Central
        geo_bin.append(2)
    elif (i<=17):
        #West_South_Central
        geo_bin.append(3)
    elif (i<=20):
        #Middle_Atlantic
        geo_bin.append(4)
    elif (i<=25):
        #New_England
        geo_bin.append(5)
    elif (i<=30):
        #East_North_Central
        geo_bin.append(6)
    elif (i<=37):
        #West_North_Central
        geo_bin.append(7)
    elif(i<=45):
        #Mountain
        geo_bin.append(8)
    else:
        #Pacific
        geo_bin.append(9)

#National 2015 population was 321 million according to US 2015 Census, police shootings during this time period was: 438
national_average = 1000000*(438/321000000)
geo_map = {1: 'South Atlantic', 2:'East South Central', 3 : 'West South Central', 4:'Middle Atlantic',\
           5:'East North Central',6:'West North Central', 7:'West North Central', 8: 'Mountain', 9:'Pacific'}
region_map = {1:'South',2:'South',3:'South',4:'North East',5:'North East',6:'Mid West',7:'Mid West',8:'West',9:'West'}
reg_num_map = {'South':1, 'North East':2,'Mid West':3,'West':4}
geo_dpm = pd.DataFrame({'Deaths by Geographic Region':geo_bin, 'State':state_bins,'Region':geo_bin,'Region Number':geo_bin, 'Deaths/million residents':state_rate})
geo_dpm['State'] = geo_dpm['State'].map(r_state)
geo_dpm['Region'] = geo_dpm['Region'].map(region_map)
geo_dpm['Region Number'] = geo_dpm['Region Number'].map(region_map)
geo_dpm['Region Number'] = geo_dpm['Region Number'].map(reg_num_map)
geo_dpm['Deaths by Geographic Region'] = geo_dpm['Deaths by Geographic Region'].map(geo_map)

plt.figure(figsize=(15,8))
ax = sns.boxplot(x="Deaths by Geographic Region", y="Deaths/million residents",hue="Region", data=geo_dpm)
ax.axhline(national_average, color='red')

ax.set_title('Police killings histograms by geographic region');
```

```python
"""Our hypothesis for Police killings by geographic region was for the Southern states to be higher, based on the presumption
or racism in those states. It is true that the Southern Region does appear to have higher rates of police killings than the North
East and the Mid West, It does not appear to have more than the West Coast, which does not have the same reputation for
intolerance."""
```
